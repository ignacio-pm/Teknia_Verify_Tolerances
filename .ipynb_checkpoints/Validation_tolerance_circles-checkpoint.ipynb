# Teknia: Validate the distance between holes in a tool

Import packages and load original image

```python
import cv2 as cv2
import numpy as np
import os
from matplotlib import pyplot as plt
from matplotlib import image as mp_image
from imutils import perspective
import math
from skimage import img_as_ubyte
from scipy import ndimage
import skimage.color as sc


os.environ["ENABLE_CV2_IMSHOW"] = 'true'

image_file = "IMG_7007.jpg"
image_file2 = "IMG_7008.jpg"
image = mp_image.imread(image_file)
image2 = mp_image.imread(image_file2)
plt.imshow(image)
plt.show()
```

Apply a Sobel filter to detect the edges of the tool in the image

```python
def edge_sobel(image):
    image = sc.rgb2gray(image) # Convert color image to gray scale
    dx = ndimage.sobel(image, 1)  # horizontal derivative
    dy = ndimage.sobel(image, 0)  # vertical derivative
    mag = np.hypot(dx, dy)  # magnitude
    mag *= 255.0 / np.amax(mag)  # normalize (Q&D)
    mag = mag.astype(np.uint8)
    return mag

# Apply the transformation and convert into OpenCV readable file
sobel_image_1 = edge_sobel(np.array(image))
cv_image_1 = img_as_ubyte(sobel_image_1)

sobel_image_2 = edge_sobel(np.array(image2))
cv_image_2 = img_as_ubyte(sobel_image_2)

# Create plot
fig = plt.figure(figsize=(16, 12))

# Plot image
a=fig.add_subplot(1, 3, 3)
image_plot_2 = plt.imshow(sobel_image_1, cmap="gray") # The image is grayscale after the transformation
a.set_title("Sobel")



plt.show()
```

Functions that allow to define the circles and the distance. Applied to an image with a single tool

```python
# Distance between two points (in this case, two centers)
def distance(p1,p2): 
    return math.sqrt((p1[0] - p1[1])**2 + (p2[0] - p2[1])**2)
  
x1=[]
y1=[]

# Main function that search, draw and calculate the distance between circles
def define_circles(cv_image):
    # Detect circles in the image. Important to adjust minRadius
    # and maxRadius to values that fit the image
    circles = cv2.HoughCircles(cv_image, cv2.HOUGH_GRADIENT, 1.7, 500,param1=170,param2=75,minRadius=0,maxRadius=30)
    output = image.copy() 

    # Ensure at least some circles were found
    if circles is not None:
      # Convert the (x, y) coordinates and radius of the circles to integers
      circle = np.round(circles[0, :]).astype("int")

      # Loop over the (x, y) coordinates and radius of the circles
      for (x, y, r) in circle:
        # Draw the circle in the output image, then draw a rectangle
        # corresponding to the center of the circle
        cv2.circle(output, (x, y), r, (0, 255, 0), 4)
        cv2.rectangle(output, (x - 5, y - 5), (x + 5, y + 5), (0, 128, 255), -1)
        x1= np.append(x1,x)
        y1= np.append(y1,y)
      print("Position of centers in X:",x1)
      print("Position of centers in Y:",y1)

      # Print 1 distance if 2 circles where detected
      if len(x1) == 2:
        dist = distance(x1,y1)
        print(dist)
      # Print 2 distances if 4 circles are detected  
      if len(x1) == 4:
        index1= np.where(y1 == np.amin(y1))
        index2= np.where(y1 == np.amin(y1[y1 != np.amin(y1)]))
        index4= np.where(y1 == np.amax(y1))
        index3= np.where(y1 == np.amax(y1[y1 != np.amax(y1)]))
        dist1 = distance([x1[index1],x1[index2]],[y1[index1],y1[index2]])
        dist2= distance([x1[index3],x1[index4]],[y1[index3],y1[index4]])
        print(dist1)
        print(dist2)

      # Show the output image
      plt.imshow(output)
    else:
      print("Nothing Found!")
    
# Apply the function to the image with one tool
define_circles(sobel_image_1)
```
